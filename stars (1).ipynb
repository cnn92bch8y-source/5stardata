```python
import numpy as np
import os
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

def process_3d_star_data(file_path):
    if not os.path.exists(file_path):
        print(f"File missing: {file_path}")
        return None, set(), []

    raw_data = np.genfromtxt(file_path, ndmin=2)
    data = raw_data[:20, :] if raw_data.shape[0] >= 20 else raw_data
    Nstar = data.shape[0]

    m = data[:, 0]
    x, y, z = data[:, 1], data[:, 2], data[:, 3]
    vx, vy, vz = data[:, 4], data[:, 5], data[:, 6]

    G = 6.67430e-11
    pair_energies = []

    for i in range(Nstar):
        for j in range(i + 1, Nstar):
            m1, m2 = m[i], m[j]

            r = np.sqrt((x[j] - x[i])**2 + (y[j] - y[i])**2 + (z[j] - z[i])**2)
            vrel_sq = (vx[j] - vx[i])**2 + (vy[j] - vy[i])**2 + (vz[j] - vz[i])**2

            mu = (m1 * m2) / (m1 + m2)
            KE = 0.5 * mu * vrel_sq
            PE = -G * m1 * m2 / r
            E = KE + PE

            pair_energies.append(((i, j), E))

    pair_energies.sort(key=lambda item: item[1])

    true_3d_binaries = set()
    seen_stars = set()

    for (i, j), E in pair_energies:
        if E < 0 and i not in seen_stars and j not in seen_stars:
            true_3d_binaries.add((i, j))
            seen_stars.update([i, j])

    return data, true_3d_binaries, pair_energies


def create_random_plane():
    n = np.random.uniform(-1.0, 1.0, 3)
    return n / np.sqrt(np.sum(n**2))


def project_3d_to_2d(x, y, z, n_hat):
    if abs(n_hat[0]) < 0.9:
        t = np.array([1.0, 0.0, 0.0])
    else:
        t = np.array([0.0, 1.0, 0.0])
        
    u = np.cross(t, n_hat)
    u = u / np.sqrt(np.sum(u**2))
    v = np.cross(n_hat, u)
    
    alpha = x * u[0] + y * u[1] + z * u[2]
    beta = x * v[0] + y * v[1] + z * v[2]
    return alpha, beta


def find_projected_binaries(alpha, beta, dmax):
    Nstar = len(alpha)

    pairs = []
    for i in range(Nstar):
        for j in range(i + 1, Nstar):
            dist = np.sqrt((alpha[i] - alpha[j])**2 + (beta[i] - beta[j])**2)
            if dist <= dmax:
                pairs.append((dist, (i, j)))
                
    pairs.sort(key=lambda item: item[0])
    
    projected_binaries = set()
    seen_stars = set()
    
    for dist, (i, j) in pairs:
        if i not in seen_stars and j not in seen_stars:
            projected_binaries.add((i, j))
            seen_stars.update([i, j])
        
    return projected_binaries


def evaluate_binary_classification(Nstar, true_3d_binaries, projected_2d_binaries):
    TP, FP, FN, TN = 0, 0, 0, 0
    tp_pairs, fp_pairs, fn_pairs = [], [], []

    for i in range(Nstar):
        for j in range(i + 1, Nstar):
            pair = (i, j)
            is_true = pair in true_3d_binaries
            is_pred = pair in projected_2d_binaries

            if is_true and is_pred:
                TP += 1
                tp_pairs.append(pair)
            elif not is_true and is_pred:
                FP += 1
                fp_pairs.append(pair)
            elif is_true and not is_pred:
                FN += 1
                fn_pairs.append(pair)
            else:
                TN += 1

    return {
        "TP": TP, "FP": FP, "FN": FN, "TN": TN,
        "tp_pairs": tp_pairs, "fp_pairs": fp_pairs, "fn_pairs": fn_pairs
    }


# NEW 2D PLOT (points only, color coded)
def plot_projected_binaries(alpha, beta, results, true_3d_binaries, projected_2d_binaries, filename=""):
    Nstar = len(alpha)

    star_colors = {}

    for i, j in results['tp_pairs']:
        star_colors[i] = 'red'
        star_colors[j] = 'red'

    for i, j in results['fp_pairs']:
        star_colors[i] = 'blue'
        star_colors[j] = 'blue'

    for i, j in results['fn_pairs']:
        star_colors[i] = 'green'
        star_colors[j] = 'green'

    plt.figure(figsize=(8, 8))

    for i in range(Nstar):
        color = star_colors.get(i, 'gray')

        plt.scatter(alpha[i], beta[i],
                    color=color,
                    edgecolors='black',
                    s=220,
                    alpha=0.9)

    x_min, x_max = np.min(alpha), np.max(alpha)
    y_min, y_max = np.min(beta), np.max(beta)
    dx = (x_max - x_min) * 0.08 if x_max != x_min else 1.0
    dy = (y_max - y_min) * 0.08 if y_max != y_min else 1.0

    plt.xlim(x_min - dx, x_max + dx)
    plt.ylim(y_min - dy, y_max + dy)

    plt.xlabel("Alpha (m)")
    plt.ylabel("Beta (m)")
    plt.title(f"Sky Map: {filename}")
    plt.grid(True, alpha=0.4)
    plt.gca().set_aspect('equal', adjustable='datalim')

    legend_elements = [
        Line2D([0], [0], marker='o', color='w', label='True Positive (TP)',
               markerfacecolor='red', markeredgecolor='black', markersize=12),
        Line2D([0], [0], marker='o', color='w', label='False Positive (FP)',
               markerfacecolor='blue', markeredgecolor='black', markersize=12),
        Line2D([0], [0], marker='o', color='w', label='False Negative (FN)',
               markerfacecolor='green', markeredgecolor='black', markersize=12),
        Line2D([0], [0], marker='o', color='w', label='True Negative (TN)',
               markerfacecolor='gray', markeredgecolor='black', markersize=12)
    ]

    plt.legend(handles=legend_elements, loc='upper right')

    plt.show()


data_files = ['./stardata_5stars.txt', './stardata.txt']
AU_TO_M = 1.495978707e11
dmax_meters = 1e4 * AU_TO_M

for file_path in data_files:
    if not os.path.exists(file_path):
        continue

    print(f"\nProcessing: {file_path}")

    data, true_3d_binaries, sorted_pair_energies = process_3d_star_data(file_path)
    Nstar = data.shape[0]

    x, y, z = data[:, 1], data[:, 2], data[:, 3]
    n_hat = create_random_plane()
    alpha, beta = project_3d_to_2d(x, y, z, n_hat)
    projected_2d_binaries = find_projected_binaries(alpha, beta, dmax=dmax_meters)

    results = evaluate_binary_classification(Nstar, true_3d_binaries, projected_2d_binaries)
    print(f"Results -> TP: {results['TP']}, FP: {results['FP']}, FN: {results['FN']}, TN: {results['TN']}")

    plot_projected_binaries(alpha, beta, results, true_3d_binaries, projected_2d_binaries, filename=os.path.basename(file_path))
```

```python
# TASK 5 3D plotting

from mpl_toolkits.mplot3d import Axes3D

def plot_3d_binaries(data, results, filename=""):
    Nstar = data.shape[0]

    x = data[:, 1]
    y = data[:, 2]
    z = data[:, 3]

    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')

    # Plot all stars first
    for i in range(Nstar):
        ax.scatter(x[i], y[i], z[i],
                   color='gray',
                   edgecolors='black',
                   s=80,
                   alpha=0.5)

    # TP = solid red
    for i, j in results['tp_pairs']:
        ax.plot3D([x[i], x[j]],
                  [y[i], y[j]],
                  [z[i], z[j]],
                  color='red',
                  linestyle='-',
                  linewidth=3,
                  label='TP' if (i, j) == results['tp_pairs'][0] else "")

        ax.scatter([x[i], x[j]],
                   [y[i], y[j]],
                   [z[i], z[j]],
                   color='red',
                   s=120)

    # FP = dashed blue
    for i, j in results['fp_pairs']:
        ax.plot3D([x[i], x[j]],
                  [y[i], y[j]],
                  [z[i], z[j]],
                  color='blue',
                  linestyle='--',
                  linewidth=3,
                  label='FP' if (i, j) == results['fp_pairs'][0] else "")

        ax.scatter([x[i], x[j]],
                   [y[i], y[j]],
                   [z[i], z[j]],
                   color='blue',
                   s=120)

    # FN = dash-dot green
    for i, j in results['fn_pairs']:
        ax.plot3D([x[i], x[j]],
                  [y[i], y[j]],
                  [z[i], z[j]],
                  color='green',
                  linestyle='-.',
                  linewidth=3,
                  label='FN' if (i, j) == results['fn_pairs'][0] else "")

        ax.scatter([x[i], x[j]],
                   [y[i], z[j]],
                   [z[i], z[j]],
                   color='green',
                   s=120)


    ax.set_xlabel('X Position')
    ax.set_ylabel('Y Position')
    ax.set_zlabel('Z Position')
    ax.set_title(f'3D Star Binary Map: {filename}')

    ax.legend(loc='upper left')

    plt.show()

plot_3d_binaries(data, results, filename=os.path.basename(file_path))
```
